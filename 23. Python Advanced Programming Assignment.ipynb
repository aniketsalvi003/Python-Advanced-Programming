1. Create a class Sudoku that takes a string as an argument. The string will contain the numbers of a regular 9x9 sudoku board left to right and top to bottom, with zeros filling up the empty cells.

Attributes

An instance of the class Sudoku will have one attribute:

   - board: a list representing the board, with sublits for each row, with the numbers as integers. Empty cell represented with 0.

Methods

An instance of the class Sudoku wil have three methods:

   - get_row(n): will return the row in position n.
   - get_col(n): will return the column in position n.
   - get_sqr([n, m]): will return the square in position n if only one argument is given, and the square to which the cell in position (n, m) belongs to if two arguments are given.

```python
import numpy as np
class Sudoku:
    def __init__(self, num_str):
        self.num_str=num_str
        
        list_=[]
        for num in self.num_str:
            list_.append(int(num))
        
        self.gameboard = np.array(list_).reshape(9,9)
        
    def get_row(self,n):
        return self.gameboard[n]
    
    def get_col(self,n):
        return self.gameboard[:,n]
    
    def get_sqr(self,n,m=0):
        square_dict={
        0:self.gameboard[:3,:3],
        1:self.gameboard[:3,3:6],
        2:self.gameboard[:3,6:9],
        3:self.gameboard[3:6,:3],
        4:self.gameboard[3:6,3:6],
        5:self.gameboard[3:6,6:9],
        6:self.gameboard[6:9,:3],
        7:self.gameboard[6:9,3:6],
        8:self.gameboard[6:9,6:9]
        }
        if m==0:
            return square_dict[n]
        else:
            if n>=0 and n<=2 and m >=0 and m<=2:
                return square_dict[0]
            elif n>=0 and n<=2 and m >=3 and m<=5:
                return square_dict[1]
            elif n>=0 and n<=2 and m >=6 and m<=8:
                return square_dict[2]
            if n>=3 and n<=5 and m >=0 and m<=2:
                return square_dict[3]
            elif n>=3 and n<=5 and m >=3 and m<=5:
                return square_dict[4]
            elif n>=3 and n<=5 and m >=6 and m<=8:
                return square_dict[5]
            if n>=6 and n<=8 and m >=0 and m<=2:
                return square_dict[0]
            elif n>=6 and n<=8 and m >=3 and m<=5:
                return square_dict[1]
            elif n>=6 and n<=8 and m >=6 and m<=8:
                return square_dict[1]
        
game = Sudoku("417950030000000700060007000050009106800600000000003400900005000000430000200701580")
print(game.gameboard)            
print(game.get_row(0)) 
print(game.get_col(8))
print(game.get_sqr(1))
print(game.get_sqr(1,8))
```

3. Write a class called CoffeeShop, which has three instance variables:

   1. name : a string (basically, of the shop)
   2. menu : a list of items (of dict type), with each item containing the item (name of the item), type (whether a food or a drink) and price.
   3. orders : an empty list

and seven methods:

  1. add_order: adds the name of the item to the end of the orders list if it exists on the menu, otherwise, return "This item is currently unavailable!"
  2. fulfill_order: if the orders list is not empty, return "The {item} is ready!". If the orders list is empty, return "All orders have been fulfilled!"
  3. list_orders: returns the item names of the orders taken, otherwise, an empty list.
  4. due_amount: returns the total amount due for the orders taken.
  5. cheapest_item: returns the name of the cheapest item on the menu.
  6. drinks_only: returns only the item names of type drink from the menu.
  7. food_only: returns only the item names of type food from the menu.

IMPORTANT: Orders are fulfilled in a FIFO (first-in, first-out) order.

Examples

tcs.add_order("hot cocoa") ➞ "This item is currently unavailable!"
# Tesha's coffee shop does not sell hot cocoa
tcs.add_order("iced tea") ➞ "This item is currently unavailable!"
# specifying the variant of "iced tea" will help the process

tcs.add_order("cinnamon roll") ➞  "Order added!"
tcs.add_order("iced coffee") ➞ "Order added!"
tcs.list_orders ➞ ["cinnamon roll", "iced coffee"]
# all items of the current order

tcs.due_amount() ➞ 2.17

tcs.fulfill_order() ➞ "The cinnamon roll is ready!"
tcs.fulfill_order() ➞ "The iced coffee is ready!"
tcs.fulfill_order() ➞ "All orders have been fulfilled!"
# all orders have been presumably served

tcs.list_orders() ➞ []
# an empty list is returned if all orders have been exhausted

tcs.due_amount() ➞ 0.0
# no new orders taken, expect a zero payable

tcs.cheapest_item() ➞ "lemonade"
tcs.drinks_only() ➞ ["orange juice", "lemonade", "cranberry juice", "pineapple juice", "lemon iced tea", "vanilla chai latte", "hot chocolate", "iced coffee"]
tcs.food_only() ➞ ["tuna sandwich", "ham and cheese sandwich", "bacon and egg", "steak", "hamburger", "cinnamon roll"]



```python
menu={"orange juice":["drink",3], "lemonade":["drink",1], "cranberry juice":["drink",4], "pineapple juice":["drink",2],
      "lemon iced tea":["drink",3],"vanilla chai latte":["drink",4], "hot chocolate":["drink",2], "iced coffee":["drink",3],
     "tuna sandwich":["food",5], "ham and cheese sandwich":["food",6], "bacon and egg":["food",7], "steak":["food",5], 
      "hamburger":["food",6], "cinnamon roll":["food",7]}
orders=[]

class CoffeeShop:
    def __init__(self, name, menu, orders):
        self.name=name
        self.menu=menu
        self.orders=orders
    
    def add_order(self, order):
        if order in self.menu.keys():
            self.orders.append(order)
            print("Order added!")
        else:
            print("This item is currently unavailable!")
            
    def fulfill_order(self):
        if len(self.orders) != 0:
            print("The", self.orders[0] , "is ready")
            del self.orders[0]
        else:
            print("All orders have been fulfilled!")
            
    def list_orders(self):
        return self.orders
    
    def due_amount(self):
        amount=0
        for order in self.orders:
            amount+=menu[order][1]
        
        return amount
    
    def cheapest_item(self):
        cheapest=["lemonade"]
        cheap=menu["lemonade"][1]
        for order in self.menu:
            if menu[order][1] < cheap:
                cheapest.append(order)
        return cheapest
    
    def drinks_only(self):
        drinks=[]
        
        for item in self.menu.keys():
            if self.menu[item][0] == "drink":
                drinks.append(item)
    
        return drinks
            
    def food_only(self):
        food=[]
        
        for item in self.menu.keys():
            if self.menu[item][0] == "food":
                food.append(item)
    
        return food
               
tcs=CoffeeShop("tcs",menu,orders)
tcs.add_order("hot cocoa") 
tcs.add_order("iced tea") 
tcs.add_order("cinnamon roll") 
tcs.add_order("iced coffee") 
print(tcs.list_orders() )
print(tcs.due_amount() )
tcs.fulfill_order() 
tcs.fulfill_order() 
tcs.fulfill_order() 
print(tcs.list_orders())
print(tcs.due_amount() )
print(tcs.cheapest_item())
print(tcs.drinks_only() )
print(tcs.food_only())
```

```python
class loneliest_number :
    def __init__(self, lo, hi):
        self.lo=lo
        self.hi=hi
        
        self.number=0
        self.distance=0
        self.closest=0
        
        self.prime_list=[]
        
        for i in range(self.lo, self.hi+1):
            if i>1:
                for j in range(2,i):
                    if(i % j==0):
                        break
                else:
                    self.prime_list.append(i)
                    
    def check(self):
        self.number=self.prime_list[0]
        
        for i in range(1,len(self.prime_list)-1):
            if (self.prime_list[i+1] - self.prime_list[i] > self.distance) or (self.prime_list[i-1]  - self.prime_list[i] > self.distance):
                self.number=self.prime_list[i]
                self.distance=max (self.prime_list[i+1] - self.prime_list[i],self.prime_list[i-1]  - self.prime_list[i])
                self.closest= self.prime_list[i+1]
            
        return { "number" : self.number , 'distance': self.distance,'closest' : self.closest }


                
l=loneliest_number(8,123)
print(l.check())
```

```python
class Selfie:
    def __init__(self,x):
        self.x=x
        self.state=[]
        
    def return_state(self):
        return self.x
    
    def save_state(self):
        self.state.append(self.x)
        print(self.state)
     
    def recover_state(self,y):
        if y >= 0 and y < len(self.state):
            self.x=self.state[y]
            return self.state[y]
        else:
            return self.x
    
    
    
p = Selfie(0)
p.x = 2
print(p.return_state())
p.save_state()
p.x = 5
p.save_state()
print(p.return_state())
p.recover_state(0)
print(p.return_state())
```
